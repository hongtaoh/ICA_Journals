```python
import pandas as pd
import urllib.parse
```

```python
df = pd.read_csv('../data/interim/author_with_pred_raw.csv')
df.head()
```

```python
df.columns
```

```python
name1 = df.authorFullName[0]
name1
```

```python
aff1 = df.affProcessed[0]
aff1
```

```python
string = 'https://www.google.com/search?q='
```

```python
string + name1 + ' ' + aff1
```

```python
string + urllib.parse.quote_plus(name1 + aff1)
```

```python
df['googleSearch'] = df.apply(
    lambda row: string + urllib.parse.quote_plus(str(row['authorFullName']) + str(row['affProcessed'])), axis=1)
```

```python
df.head()
```

```python
df['authorID'] = df['doi'] + '+' + df['authorPosition'].astype(str)
```

```python
df['authorID'][0]
```

```python
df['googleSearch'].to_csv('google1.csv', index = False)
```
