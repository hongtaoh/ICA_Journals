## Aim

To see how many papers/authors belong to research papers. 

```python
import pandas as pd
```

```python
df = pd.read_csv('cat-issueurl.csv')
```

```python
df = df[['category.1', 'issueURL.1', 'TO_EXCLUDE']]
df
```

```python
exclude_dic = dict(zip(df['category.1'], df.TO_EXCLUDE))
exclude_dic
```

```python
exclude_dic['Reviews']
```

```python
paper = pd.read_csv('../data/interim/ica_paper_df.csv')
```

```python
paper.head()
```

```python
paper['category']
```

```python
paper['to_exclude'] = paper['category'].apply(lambda x: exclude_dic[x])
```

```python
paper.shape
```

```python
research_paper = paper[paper.to_exclude == False].reset_index(drop=True)
research_paper.tail()
```

```python
research_paper.shape
```

```python
research_paper_dois = research_paper.doi
research_paper_dois
```

```python
author = pd.read_csv('../data/interim/ica_author_data.csv')
author.head()
```

```python
author.shape
```

```python
authors_to_keep = author[author.doi.isin(research_paper_dois)]
authors_to_keep.shape
```
