```python
import pandas as pd
import numpy as np
import re
```

```python
cat_df = pd.read_csv('../data/raw/category-classification.csv')[['category', 'issueURL', 'TO_EXCLUDE']]
cat_df
```

```python
cat_df = cat_df[['category', 'issueURL', 'TO_EXCLUDE']]
cat_df.head()
```

```python
to_exclude_cat = cat_df[cat_df.TO_EXCLUDE == True]['category']
```

```python
len(to_exclude_cat)
```

```python
to_exclude_cat.head(10)
```

```python
to_delete_dois_df = pd.read_csv('../data/raw/to_delete_dois.txt', header=None).iloc[:, 0]
to_delete_dois = [re.sub('https://doi.org/', '', x) for x in to_delete_dois]
to_delete_dois
```

```python
papers = pd.read_csv('../data/interim/ica_paper_df.csv')
papers.head()
```

```python
papers['to_exclude'] = np.where(
    (papers.category.isin(to_exclude_cat)) | (papers.doi.isin(to_delete_dois)), 
    True, 
    False)
```

```python
papers.shape
```

```python
research_papers = papers[papers.to_exclude == False]
research_papers.shape
```

```python
research_papers.head()
```

```python
research_papers_dois = research_papers.doi
```
