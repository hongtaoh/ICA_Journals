```python
paper['to_exclude'] = paper['category'].apply(lambda x: exclude_dic[x])
```

```python
paper.shape
```

```python
research_paper = paper[paper.to_exclude == False].reset_index(drop=True)
research_paper.tail()
```

```python
research_paper.shape
```

```python
research_paper_dois = research_paper.doi
research_paper_dois
```

```python
to_delete_dois = pd.read_csv('../data/raw/to_delete_dois.txt', header=None)
to_delete_dois.columns = ['doi']
to_delete_dois
```

```python
research_paper_dois = [x for x in research_paper_dois if x not in to_delete_dois]
```

```python
len(research_paper_dois)
```

```python
author = pd.read_csv('../data/interim/ica_author_data.csv')
author.head()
```

```python
authors_to_keep = author[author.doi.isin(research_paper_dois)]
authors_to_keep.shape
```
