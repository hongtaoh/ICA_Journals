## Aim

The aim of this notebook is to randomly select 113 names from the final author dataset. These 113 names will be used to see how reliant we are on api results. 

```python
import pandas as pd
import numpy as np
```

```python
# this is the final author set. I brought it hear mainly to get the list of authorID
final_authors = pd.read_csv('../../data/processed/authors_to_study_expanded.csv')
```

```python
final_authors.shape
```

```python
final_authors_id = final_authors.authorID.tolist()
```

```python
# this is the 11304 authors dataset. It contains all the information we used for gender & race coding
# (and icr calculation), so it's best that I brought it here first and select the final authorID rows
authors = pd.read_csv('../../data/processed/authors_to_study.csv')
```

```python
authors_to_sample_from = authors[authors.authorID.isin(final_authors_id)]
```

```python
authors_to_sample_from.head(2)
```

```python
random_113 = authors_to_sample_from.sample(113, random_state = 2023)
```

```python
random_113.head(2)
```

```python
random_113.to_csv('../../data/interim/api_reliance_checking/113_names_for_api_reliance_checking.csv')
```
